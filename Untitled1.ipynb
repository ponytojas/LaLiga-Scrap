```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
website = requests.get('https://www.bdfutbol.com/es/t/t2019-201.html').text
soup = BeautifulSoup(website,'lxml')
```

```python
soup
```

```python
players = soup.find('table',{'id':'taulaplantilla'})
```

```python
players
```

```python
playersTemp = players.findAll('span', {'class': 'd-none'})
```

```python
playersTemp
```

```python
for player in playersTemp:
    playerName = str(player)
    result = re.search('">(.*)</' ,playerName)
    result = result.group(1)
```
